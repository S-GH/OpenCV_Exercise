#### img slice
* slice = img[y : y+h, x : x+w]

```python
# 이미지 slice
img = cv.imread('img/sunset.jpg')
y = 100; y_end = y+20;
x = 100; x_end = x+20;
img[y:y_end] = (0,0,255)
plt.imshow(cv.cvtColor(img,cv.COLOR_BGR2RGB))
plt.show()

img[:,x:x_end] = (255,0,0)
plt.imshow(cv.cvtColor(img,cv.COLOR_BGR2RGB))
plt.show()
```

### ROI 지정 (관심영역)

```python
img = cv.imread('img/sunset.jpg')

x=320; y=150; w=50; h=50
img_ = cv.rectangle(img, (x,y),(x+w,y+h),(0,0,255),2)
plt.imshow(cv.cvtColor(img_,cv.COLOR_BGR2RGB))
plt.show()

img__ = img[y:y+h,x:x+w]
plt.imshow(cv.cvtColor(img__,cv.COLOR_BGR2RGB))
plt.show()

# slice를 활용해 특정 좌표에 특정 그림을 넣는다
img[y:y+h,x+w:x+(2*w)] = img[y:y+h,x:x+w]
img2 = cv.rectangle(img, (x,y),(x+w,y+h),(0,0,255),2)
img2 = cv.rectangle(img, (x+w,y),(x+(2*w),y+h),(0,0,255),2)
plt.imshow(cv.cvtColor(img2,cv.COLOR_BGR2RGB))
plt.show()
```

```python
## 마우스 이벤트로 관심영역 지정
isDragging = False                      # 마우스 드래그 상태 저장 
x0, y0, w, h = -1,-1,-1,-1              # 영역 선택 좌표 저장
blue, red = (255,0,0),(0,0,255)         # 색상 값 

def mouse_callback(event,x,y,flags,param):
    global isDragging, x0, y0, img # 전역변수 끌어다쓰기
    if event == cv.EVENT_LBUTTONDOWN:  # 왼쪽 마우스 버튼 다운, 드래그 시작
        isDragging = True
        x0 = x
        y0 = y
    elif event == cv.EVENT_MOUSEMOVE:  # 마우스 움직임=
        if isDragging:                  # 드래그 진행 중
            img_draw = img.copy()       # 사각형 그림 표현을 위한 이미지 복제
            cv.rectangle(img_draw, (x0, y0), (x, y), blue, 2) # 드래그 진행 영역 표시
            cv.imshow('img', img_draw) # 사각형 표시된 그림 화면 출력
    elif event == cv.EVENT_LBUTTONUP:  # 왼쪽 마우스 버튼 업
        if isDragging:                  # 드래그 중지
            isDragging = False          
            w = x - x0                  # 드래그 영역 폭 계산
            h = y - y0                  # 드래그 영역 높이 계산
            print(f'x:{x0}, y{y0}, w:{w}, h:{h}')
            if w > 0 and h > 0:         # 폭과 높이가 음수이면 드래그 방향이 옳음
                img_draw = img.copy()   # 선택 영역에 사각형 그림을 표시할 이미지 복제
                # 선택 영역에 빨간 사각형 표시
                cv.rectangle(img_draw, (x0, y0), (x, y), red, 2) 
                cv.imshow('img', img_draw) # 빨간 사각형 그려진 이미지 화면 출력
                roi = img[y0:y0+h, x0:x0+w] # 원본 이미지에서 선택 영영만 ROI로 지정
                cv.imshow('cropped', roi)  # ROI 지정 영역을 새창으로 표시
                cv.moveWindow('cropped', 0, 0) # 새창을 화면 좌측 상단에 이동
                cv.imwrite('./cropped.jpg', roi)   # ROI 영역만 파일로 저장
                print("croped.")
            elif w < 0 and h < 0:
                img_draw = img.copy() 
                cv.rectangle(img_draw, (x0, y0), (x, y), red, 2) 
                cv.imshow('img', img_draw)
                # 선택 영역이 -이기때문에 
                roi = img[y0+h:y0, x0+w:x0]
                cv.imshow('cropped', roi)
                cv.moveWindow('cropped', 0, 0)
                cv.imwrite('./cropped.jpg', roi)
                print("croped.")
            else:
                cv2.imshow('img', img)  # 드래그 방향이 잘못된 경우 사각형 그림ㅇㅣ 없는 원본 이미지 출력
                print("좌측 상단에서 우측 하단으로 영역을 드래그 하세요.")

img = cv.imread('img/sunset.jpg')
cv.imshow('img', img)
cv.setMouseCallback('img', mouse_callback) # 마우스 이벤트 등록
cv.waitKey()
cv.destroyAllWindows()
```

### 여러 컬러 스페이스 다루기
- RGBA(alpha=배경투명도), HSV( H:색조, S:채도, V:명도)

```python
img = cv.imread('img/smile.png')
bgr = cv.imread('img/smile.png',cv.IMREAD_COLOR)
bgra = cv.imread('img/smile.png',cv.IMREAD_UNCHANGED)
print('default shape',img.shape,'color shape',bgr.shape,'bgra shape',bgra.shape)
# default
plt.figure(figsize=(2, 2))
plt.imshow(cv.cvtColor(img,cv.COLOR_BGR2RGB))
plt.show()
# bgr space
plt.figure(figsize=(2, 2))
plt.imshow(cv.cvtColor(bgr,cv.COLOR_BGR2RGB))
plt.show()
# bgra space
plt.figure(figsize=(2, 2))
plt.imshow(cv.cvtColor(bgra,cv.COLOR_BGR2RGB))
plt.show()

# 특정 RGB값중 R값 0으로 만들어 출력
# BGR이니 0번째 인덱스
img[:,:,0] = 0
plt.figure(figsize=(2, 2))
plt.imshow(cv.cvtColor(img,cv.COLOR_BGR2RGB))
plt.show()
```

### threshold 영상 이진화


**threshold**
* ret, out = cv2.threshold(img, threshold, value, type_flag) 
    * ret : 스레시홀딩에 사용한 경계값 
    * out : 결과 바이너리 이미지
    * cv2.THRESH_BINARY : 픽셀 값이 경계 값을 넘으면 value 값을 지정하고 넘지 못하면 0 지정 
    * cv2.THRESH_BINARY_INV : ‘cv2.THRESH_BINARY ’의 반대 
    * cv2.THRESH_TRUNC : 픽셀 값이 경계 값을 넘으면 value 값을 지정하고 넘지 못하면 유지
    * cv2.THRESH_TOZERO : 픽셀 값이 경계값을 넘으면 유지, 아니면 0 
    * cv2.THRESH_TOZERO_INV : ‘cv2.THRESH_TOZERO’의 반대 

**adaptiveThreshold**
* cv2.adaptiveThreshold(img, value, method, type_flag, block_size, C)  
    * method : 경계 값 결정 방법 
        * cv2.ADAPTIVE_THRESH_MEAN_C : 이웃 픽셀의 평균으로 설정
        * cv2.ADAPTIVE_THRESH_GAUSSIAN_C : 가우시안 분포에 따른 가중치의 합으로 결정  
    * blosck_size : 영역으로 나눌 이웃의 크기 (nxn), 홀수(3,5,7,…)
    * C : 계산된 경계 값 결과에서 가감할 상수( 음수가능) 

```python
img = cv.imread('ball.png',cv2.IMREAD_GRAYSCALE)
# 127~255까지의 픽셀들을 255로 이하의값들을 0으로
ret, out = cv.threshold(img,127,255,cv.THRESH_BINARY)
plt.figure(figsize=(3, 3))
plt.imshow(out,cmap='gray')
plt.show()

# 100~255
ret, out = cv.threshold(img,100,255,cv.THRESH_BINARY)
plt.figure(figsize=(3, 3))
plt.imshow(out,cmap='gray')
plt.show()

# AdaThreshold
out = cv2.adaptiveThreshold(img, 255, cv.ADAPTIVE_THRESH_MEAN_C, cv.THRESH_BINARY, 21, 5) 
plt.figure(figsize=(3, 3))
plt.imshow(out,cmap='gray')
plt.show()

# 오츠 알고리즘
ret, t_img = cv2.threshold(img, 0, 255, cv.THRESH_BINARY | cv2.THRESH_OTSU)
plt.figure(figsize=(3, 3))
plt.imshow(t_img,cmap='gray')
plt.show()
```

### image resize, add
* cv.resize(image, (width, height), fx, fy, interpolation) : 이미지의 크기를 조절합니다. 
    * fx : 가로 비율
    * fy : 세로 비율 
    * interpolation : 보간법 
        * 보간법 : 사이즈가변할때 값을 잃는 픽셀들이 생길수있음, 그런픽셀들을 주변값을 참조해 채워주는 방법
        * INTER_CUBIC : 사이즈를 크게 할 때 주로 사용합니다. 
        * INTER_AREA : 사이즈를 작게 할 때 주로 사용합니다. 
* cv.add(src1, src2[, dest, mask, dtype]): src1과 src2 더하기 

```python
# image resize
img = cv.imread('img/smile.png')
plt.imshow(cv.cvtColor(img,cv.COLOR_BGR2RGB))
plt.show()

img_ = cv.resize(img,(100,100),0.5,0.5)
plt.imshow(cv.cvtColor(img_,cv.COLOR_BGR2RGB))
plt.show()

# 0.5배 + 보간법 사용
img_ = cv.resize(img,None,fx=0.5,fy=0.5,interpolation=cv.INTER_AREA)
plt.imshow(cv.cvtColor(img_,cv.COLOR_BGR2RGB))
plt.show()
```
