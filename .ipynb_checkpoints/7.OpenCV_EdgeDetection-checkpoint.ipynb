```python
import cv2 as cv
import numpy as np
from matplotlib import pyplot as plt
```

### EdgeDetection
- Sobel filter, Laplacian filter, canny
- contour 검출

**Sobel filter**

**cv2.Sobel(이미지,정밀도,x미분,y미분,커널,배율,델타,속성)**

```python
img = cv.imread('img/eagle.jpg')
gray = cv.cvtColor(img, cv.COLOR_BGR2GRAY)
# 소벨 필터링
fil_ = cv.Sobel(gray, -1, 1, 0, ksize = 5)
plt.imshow(cv.cvtColor(fil_, cv.COLOR_BGR2RGB)),plt.show()
```

**Laplacian filter**

**cv.Laplacian(이미지, 정밀도, 커널, 배율, 델타)**

```python
img = cv.imread('img/eagle.jpg')
gray = cv.cvtColor(img, cv.COLOR_BGR2GRAY)
# 라플라시안 필터링
fil_ = cv.Laplacian(gray, cv.CV_8U, ksize=3)
plt.imshow(cv.cvtColor(fil_, cv.COLOR_BGR2RGB)),plt.show()
```

**Canny**

**cv.Canny(이미지,임계값1,임계값2,커널크기,L2 or L1)**

```python
img = cv.imread('img/eagle.jpg')
gray = cv.cvtColor(img, cv.COLOR_BGR2GRAY)
# 라플라시안 필터링
fil_ = cv.Canny(gray,100,255)
plt.imshow(cv.cvtColor(fil_, cv.COLOR_BGR2RGB)),plt.show()
```

## CONTOUR DETECT

- findCountours / drawContours / convexHull / approxPolyDP
- moments (컨투어 특징추출) / contourArea (컨투어 면적) / arcLength (컨투어 둘레)
- grabCut (오브젝트가 아닌 배경을 추출하는 알고리즘)

**img, contours, hiearchy = cv.findContours(image, mode, method)**
- hiearchy(계층트리) : 컨투어 * 4의 행렬
    * 1번 원소: 같은 수준의 다음 컨투어의 인덱스. 같은 수준의 다음 컨투어가 없으면 -1
    * 2번 원소: 같은 수준의 이전 컨투어의 인덱스. 같은 수준의 이전 컨투어가 없으면 -1
    * 3번 원소: 하위 자식 컨투어의 인덱스. 가장 하위의 컨투어면 -1
    * 4번 원소: 부모 컨투어의 인덱스. 가장 상위의 컨투어면 -1
- mode : 컨투어 제공방식 
	* cv2.RETR_EXTERNAL : 가장 바깥쪽 라인만 제공
	* cv2.RETR_LIST : 모든 라인을 계층 없이 제공
	* cv2.RETR_CCOMP:모든 라인을 2계층으로 제공
	* cv2.RETR_TREE : 모든 라인의 모든 계층 정보를 트리 구조로 제공
- method : 근사 값 방식 선택 
	* cv2.CHAIN_APPROX_NONE : 근사 계산하지 않고 모든 좌표 제공
	* cv2.CHAIN_APPROX_SIMPLE : 컨투어 꼭짓점 좌표만 제공
	*  cv2.CHAIN_APPROX_TC89_L1 :The-Chin 알고리즘으로 좌표 개수 축소
	*  cv2.CHAIN_APPROX_TC89_KCOS :The-Chin 알고리즘으로 좌표 개수 축소

**cv.drawContours(img, contours, contour_Idx, color, thickness)**
- contourIdx : 그림 그릴 컨투어 인덱스 ( -1 : 모든 컨투어 표시 ) 
- color : 색상 값 

```python
# 컨투어 검출 절차 : BGR > GRAY > Binary > findContour > drawContour
img_ = cv.imread('img/contour2.png')
img = cv.cvtColor(img_,cv.COLOR_BGR2GRAY)
ret, thresh = cv.threshold(img,230,255,0)
thresh = cv.bitwise_not(thresh)

contours, hierachy = cv.findContours(thresh, cv.RETR_TREE,cv.CHAIN_APPROX_SIMPLE)
img_ = cv.drawContours(img_, contours, -1, (0,0,255), 4) # -1 : 모든 외곽
plt.imshow(cv.cvtColor(img_, cv.COLOR_BGR2RGB)),plt.show()

# 컨투어 숫자, 넓이, 길이, 중심점 출력
cnt = 0
for i in contours :
    cnt += 1
    # 컨투어 특징 추출
    mmt = cv.moments(i)
    # m10/m00, m01/m00  중심점 계산
    cx = int(mmt['m10']/mmt['m00'])
    cy = int(mmt['m01']/mmt['m00'])    
    sp = mmt['m00'] # 넓이
    length = cv.arcLength(c, True) # 컨투어 외곽선 길이
    
    # 컨투어 인덱스 그리기
    cv.putText(img_,str(cnt),(i[0][0][0],i[0][0][1]),cv.FONT_HERSHEY_SCRIPT_SIMPLEX,0.7,(255,0,0),2)
    # 중심점
    cv.circle(img_, (cx, cy), 5, (0, 255, 255), -1)
    # 넓이
    cv.putText(img_, f"s:{sp}", (cx, cy+20) , cv.FONT_HERSHEY_PLAIN, 1, (0,0,255))
    # 컨투어 시작점에 길이 그리기
    cv.putText(img_, f"  L:{length}", tuple(i[0][0]), cv.FONT_HERSHEY_PLAIN, 1, (255,0,0))
                                                            
plt.figure(figsize=(8,8)),plt.imshow(cv.cvtColor(img_, cv.COLOR_BGR2RGB)),plt.show()
```

- convexHull : 블록의 껍질을 contour
- approxPolyDP : 컨투어의 숫자를 줄이고 실제 컨투어 라인과 근사한 라인을 그릴때 사용

```python
# 이미지를 읽어서 그레이 스케일 및 스레시홀드 변환
img_ = cv.imread('img/contour2.png')
img_c = img_.copy()
img = cv.cvtColor(img_,cv.COLOR_BGR2GRAY)
ret, thresh = cv.threshold(img,230,255,0)
thresh = cv.bitwise_not(thresh)
contours, hierachy = cv.findContours(thresh, cv.RETR_TREE,cv.CHAIN_APPROX_SIMPLE)

# 원본이미지에 그려야한다.
img_ = cv.drawContours(img_, contours, -1, (0,0,255), 4) # -1 : 모든 외곽
plt.imshow(cv.cvtColor(img_, cv.COLOR_BGR2RGB)),plt.show()

for cnt in contours:
    # 파라미터, 근사치의 최대거리 ex) 0.01 ~ 0.05
    epsilon = 0.05 * cv.arcLength(cnt, True)
    # 입력값 : 컨투어배열, 실제 컨투어 라인과 근사치의 최대거리, 폐곡선여부
    approx = cv.approxPolyDP(cnt, epsilon, True)
    # 꼭짓점 갯수
    print(len(approx))
    cv.drawContours(img_c,[approx],0,(255,0,0),4)

plt.imshow(cv.cvtColor(img_c, cv.COLOR_BGR2RGB)),plt.show()
```

```python
import cv2 as cv
import numpy as np
from matplotlib import pyplot as plt

# 이미지 읽기
img = cv.imread("img2/paper.jpg")
img_g = cv.cvtColor(img,cv.COLOR_BGR2GRAY)
img_b = cv.GaussianBlur(img_g,(5,5),0)
img_c = cv.Canny(img_b,100,255)
ret, thres = cv.threshold(img_c,176,255,0)
# thres = cv.bitwise_not(thres)

cv.imshow('a',thres)
cv.waitKey()
cv.destroyAllWindows()

_, contours, hierachy = cv.findContours(thres, cv.RETR_TREE,cv.CHAIN_APPROX_SIMPLE)
img_ = cv.drawContours(img, contours, -1, (0,0,255), 4) # -1 : 모든 외곽
plt.imshow(cv.cvtColor(img_, cv.COLOR_BGR2RGB)),plt.show()

lst = []
# find largest contour
for i in contours :
    # 컨투어 특징 추출
    mmt = cv.moments(i)
    # m10/m00, m01/m00  중심점 계산 
    sp = mmt['m00'] # 넓이
    lst.append(sp)

mx = max(lst)
large_index = lst.index(mx)
print(large_index)
print(len(contours))

epsilon = 0.01 * cv.arcLength(contours[large_index], True)
approx = cv.approxPolyDP(contours[large_index], epsilon, True)
cv.drawContours(img,[approx],0,(255,0,0),4)

approx = np.resize(approx,(4,2))
approx = approx.astype(np.float32)
nparr = np.array([approx[0],approx[1],approx[3],approx[2]])

plt.figure(figsize=(10,10))
plt.imshow(cv.cvtColor(img, cv.COLOR_BGR2RGB)),plt.show()


###### 원근법 이용해 translation ########
img_copy = cv.imread("img2/paper.jpg")
rows, cols, ch = img_copy.shape
# print(img_copy.shape)

# 좌표점은 좌상->좌하->우상->우하
pts1 = nparr
# 이동하는 점
pts2 = np.float32([[0,0], [0,rows], [cols, 0], [cols,rows]])
# print(pts2)

print(rows)
# # # pts1의 좌표에 표시. perspective 변환 후 이동 점 확인.
cv.circle(img_copy, (0,0), 5, (255,0,0),-1)
cv.circle(img_copy, (0,rows), 5, (0,255,0),-1)
cv.circle(img_copy, (cols,0), 5, (0,0,255),-1)
cv.circle(img_copy, (cols,rows), 5, (0,0,255),-1)


# # # pts1의 좌표에 표시. perspective 변환 후 이동 점 확인.
cv.circle(img_copy, (int(approx[0][0]),int(approx[0][1])), 5, (255,0,0),-1)
cv.circle(img_copy, (int(approx[1][0]),int(approx[1][1])), 5, (0,255,0),-1)
cv.circle(img_copy, (int(approx[2][0]),int(approx[2][1])), 5, (0,0,255),-1)
cv.circle(img_copy, (int(approx[3][0]),int(approx[3][1])), 5, (0,255,255),-1)

# # 4개의 점을이용
M = cv.getPerspectiveTransform(pts1, pts2)
dst = cv.warpPerspective(img_copy, M, (568,782))

plt.figure(figsize=(10,10))
plt.imshow(cv.cvtColor(img_copy, cv.COLOR_BGR2RGB)),plt.title('image'),plt.show()
plt.figure(figsize=(10,10))
plt.imshow(cv.cvtColor(dst, cv.COLOR_BGR2RGB)),plt.title('Perspective'),plt.show()
```
