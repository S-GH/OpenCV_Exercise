```python
import cv2 as cv
import numpy as np
from matplotlib import pyplot as plt
```

### Filtering
- filter2d, blur, GaussianBlur, medianBlur, BilateralFilter
- Convolutional   
    - Convloutional : n*n        .
    


**filter2d**  

```python
img = cv.imread('img/Amsterdam.jpg')
#   3*3
k = np.ones((5,5))/5**2
b = cv.filter2D(img,-1,k)
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB)),plt.show()
plt.imshow(cv2.cvtColor(b, cv2.COLOR_BGR2RGB)),plt.show()
```

**blur**   = filter2d

```python
img = cv.imread('img/Amsterdam.jpg')
# 5*5   
b = cv.blur(img,(5,5))
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB)),plt.show()
plt.imshow(cv2.cvtColor(b, cv2.COLOR_BGR2RGB)),plt.show()
```

**Gaussian Blurring**

    ,            

```python
img = cv.imread('img/Amsterdam.jpg')
# 5*5   
b = cv.GaussianBlur(img,(5,5),0)
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB)),plt.show()
plt.imshow(cv2.cvtColor(b, cv2.COLOR_BGR2RGB)),plt.show()
```

**medianBlur**

      ,   

```python
img = cv.imread('img/noise.png')
b = cv.medianBlur(img,5)
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB)),plt.show()
plt.imshow(cv2.cvtColor(b, cv2.COLOR_BGR2RGB)),plt.show()
```
