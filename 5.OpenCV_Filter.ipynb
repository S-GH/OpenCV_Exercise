```python
import cv2 as cv
import numpy as np
from matplotlib import pyplot as plt
```

### Filtering
- filter2d, blur, GaussianBlur, medianBlur, BilateralFilter
- Convolutional 계산을 통해 필터링
    - Convloutional : n*n의 커널을 이용해 해당 커널값들의 픽셀곱들을 합친 값을 넣는다.
    


**filter2d** 평균 필터

```python
img = cv.imread('img/Amsterdam.jpg')
# 커널 생성 3*3
k = np.ones((5,5))/5**2
b = cv.filter2D(img,-1,k)
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB)),plt.show()
plt.imshow(cv2.cvtColor(b, cv2.COLOR_BGR2RGB)),plt.show()
```

**blur** 평균 블러링 = filter2d

```python
img = cv.imread('img/Amsterdam.jpg')
# 5*5의 커널로 평균 필터
b = cv.blur(img,(5,5))
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB)),plt.show()
plt.imshow(cv2.cvtColor(b, cv2.COLOR_BGR2RGB)),plt.show()
```

**Gaussian Blurring**

모든 픽셀에 똑같은 가중치가 아니라, 중심에있는 픽셀에 높은 가중치를 적용해 중앙에 있는 값이 더 중요하게 작동하는 블러링

```python
img = cv.imread('img/Amsterdam.jpg')
# 5*5의 커널로 평균 필터
b = cv.GaussianBlur(img,(5,5),0)
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB)),plt.show()
plt.imshow(cv2.cvtColor(b, cv2.COLOR_BGR2RGB)),plt.show()
```

**medianBlur**

커널 영역에서 중간의 값을 대상 픽셀로 선택,무작위 노이즈를 제거하는데 좋음

```python
img = cv.imread('img/noise.png')
b = cv.medianBlur(img,5)
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB)),plt.show()
plt.imshow(cv2.cvtColor(b, cv2.COLOR_BGR2RGB)),plt.show()
```

**Bilateral Filtering**
    - 에너지를 보존하고 노이즈를 감소시킬수 있음.
    - 블러링 필터의 경계도 흐릿하게 만드는 문제를 보완
    - 가우시안 필터와 경계 필터 2개 사용
    - 노이즈는 없고 경계가 비교적 또렷한 영상을 얻음 
    - 속도가 느림
* dst = cv2.bilateralFilter(src, d, sigmaColor, sigmaSpace[, dst, borderType])
    - d : 필터의 직경
    - sigmaColor : 색공간 필터의 시그마 값
    - sigmaSpace : 좌표 공간의 시그마값 (sigmaSpace와 같은 값을 권장하며, 10~150 사이 값을 권장)

```python
img = cv.imread('img/texture.png')
# 5*5의 커널로 평균 필터
b = cv.bilateralFilter(img,9,75,75)
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB)),plt.show()
plt.imshow(cv2.cvtColor(b, cv2.COLOR_BGR2RGB)),plt.show()
```
