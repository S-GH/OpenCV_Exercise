```python
import cv2 as cv
import numpy as np
from matplotlib import pyplot as plt
```

### Morphology  
-  Erosion , Dilation , Opening , Closing 

(3*3  )


**Erosion  **

cv.erode  ,   

```python
img = cv.imread('img/morph_dot.png')
# original
plt.subplot(131),plt.imshow(cv.cvtColor(img, cv.COLOR_BGR2RGB))
# 3*3 
kernel = np.ones((3, 3), np.uint8)
result = cv.erode(img, kernel, iterations = 1)
plt.subplot(132),plt.imshow(cv.cvtColor(result, cv.COLOR_BGR2RGB))
# 6*6 
kernel = np.ones((6, 6), np.uint8)
result = cv.erode(img, kernel, iterations = 1)
plt.subplot(133),plt.imshow(cv.cvtColor(result, cv.COLOR_BGR2RGB)),plt.show()
```

**Dilation  **

cv.dilate,    ,  ,      

```python
img = cv.imread('img/morph_hole.png')
# original
plt.subplot(131),plt.imshow(cv.cvtColor(img, cv.COLOR_BGR2RGB))

# 3*3 
kernel = np.ones((3, 3), np.uint8)
# dilate !
result = cv.dilate(img, kernel, iterations = 1)
plt.subplot(132),plt.imshow(cv.cvtColor(result, cv.COLOR_BGR2RGB))

# 6*6 
kernel = np.ones((6, 6), np.uint8)
result = cv.dilate(img, kernel, iterations = 1)
plt.subplot(133),plt.imshow(cv.cvtColor(result, cv.COLOR_BGR2RGB)),plt.show()
```

**Opening  **

cv.morphologyEx, Erosion > Dilation   

      .

```python
img = cv.imread('img/morph_noise.png')
# original
plt.subplot(131),plt.imshow(cv.cvtColor(img, cv.COLOR_BGR2RGB))

# 3*3 
kernel = np.ones((3, 3), np.uint8)
result = cv.morphologyEx(img, cv.MORPH_OPEN, kernel)
plt.subplot(132),plt.imshow(cv.cvtColor(result, cv.COLOR_BGR2RGB))

# 6*6 
kernel = np.ones((6, 6), np.uint8)
result = cv.morphologyEx(img, cv.MORPH_OPEN, kernel)
plt.subplot(133),plt.imshow(cv.cvtColor(result, cv.COLOR_BGR2RGB)),plt.show()
```

**Closing  **

    .

```python
img = cv.imread('img/human_binary.png')
# original
plt.subplot(131),plt.imshow(cv.cvtColor(img, cv.COLOR_BGR2RGB))

# 3*3 
kernel = np.ones((3, 3), np.uint8)
result = cv.morphologyEx(img, cv.MORPH_CLOSE, kernel)
plt.subplot(132),plt.imshow(cv.cvtColor(result, cv.COLOR_BGR2RGB))

# 6*6 
kernel = np.ones((6, 6), np.uint8)
result = cv.morphologyEx(img, cv.MORPH_CLOSE, kernel)
plt.subplot(133),plt.imshow(cv.cvtColor(result, cv.COLOR_BGR2RGB)),plt.show()
```
